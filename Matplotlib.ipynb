___________**Plot in Matplotlib**___________

```python
%matplotlib inline
from matplotlib import pyplot as plt 
plt.plot([1,2,3,4],[1,2,4,3])
plt.title("Experiment Data")
plt.xlabel("Time ________T")
plt.ylabel("Distance ________X")
plt.show()

```

```python
%matplotlib inline
from matplotlib import pyplot as plt 

a=[1,3,5,4]
b=[10,20,18,19]
c=[2,3,4,6]
d=[5,7,6,8]
plt.title("Experiment")
plt.plot(a,b)
plt.plot(c,d)
plt.show()
```

_________**Using Styles in matplotlib**_________

```python
%matplotlib inline
from matplotlib import pyplot as plt
from matplotlib import style

style.use('bmh')
a=[1,3,5,4]
b=[10,20,18,19]
c=[2,3,4,6]
d=[5,7,6,8]

plt.title("Experiment")
plt.xlabel("t________Time")
plt.ylabel("x________Distance")
plt.plot(a,b, label="Exp1")
plt.plot(c,d, label="Exp2")
plt.legend()
plt.show()
```

______**Changing axis values**________        
______**Changing plot to others**_____

```python
%matplotlib inline
from matplotlib import pyplot as plt
from matplotlib import style

style.use('fivethirtyeight')
a=['a','b','c','d']
b=[8,15,10,19]
c=['a','b','c','f']
d=[6,7,13,9]
plt.title("Experiment")
plt.xlabel("t________Time")
plt.ylabel("x________Distance")
plt.scatter(a,b, label="Exp1")
plt.scatter(c,d, label="Exp2")
plt.legend()
plt.show()
```
